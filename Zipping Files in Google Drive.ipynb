**Adding a Zip File To Gooel Drive**

```python
from zipfile import ZipFile
import os
```

```python
file_paths = []

for root, directories, files in os.walk('/content/drive/Shareddrives/JustaStroage/Torrent/DriverPack'):
  for filename in files:
    filepath = os.path.join(root, filename)
    file_paths.append(filepath)

    with ZipFile('/content/drive/Shareddrives/JustaStroage/DriverPackSolution22.zip','w') as zip:
      for file in file_paths:
        zip.write(file)
```
